# Bloc fonctions

```python
def colormapArray(list2plot):
    '''
    Prend en argument ce qu'il y a à plotter (list2plot) pour calculer sa longueur et le pas pour slicer 
    la colormap (ici summer)
    Retourne colors : la colormap slicée
    '''
    pas=1/len(list2plot)
    list4slice=[pas*i for i in range(len(list2plot)+1)]
    colors=plt.cm.summer(list4slice)
    return colors

def infos(df,namecol):
    '''
    Prend en argument un dataframe (df) et le nom de la colonne qui nous intéresse (namecol).
    Calcule la Data serie avec les fréquences.
    Renvoie des infos, les fréquences, et les fréquences en pourcentage.
    '''
    print("Informations à propos de la colonne :", namecol)
    freq = df[namecol].value_counts()
    
    maximum = freq.max()
    print("Le plus représenté :", freq[freq == maximum].index[0])
    
    liste1 = freq[freq == 1].index.tolist()
    print("Représentants uniques :")
    print(liste1)
    
    prop = df[namecol].value_counts(normalize=True)
    prop = prop.apply(lambda x: round(100*x,2))
    print("Proportion du plus représenté :", prop.max(), "%")
    
    freq.sort_values(ascending=False, inplace=True)
    prop.sort_values(ascending=False, inplace=True)
    
    
    return freq, prop

def LONLATfromJSONlist(list_of_json):
    ret = {}
    ret['LAT'] = []
    ret['LON'] = []    
    for j in list_of_json:
        if type(j) == str:
            js = json.loads(j)
            ret["LAT"].append(js['coordinates'][1])
            ret["LON"].append(js['coordinates'][0])
        else:
            ret['LON'].append('NA')
            ret['LAT'].append('NA')    
    return(ret)

def haversine(p1,p2):
    '''
    Prend en argument 2 points p1 et p2 de type : (latitude, longitude) et retourne la distance entre les 2 en km
    '''
    # conversion en radians 
    lat1 = radians(p1[0])
    lon1 = radians(p1[1])
    lat2 = radians(p2[0])
    lon2 = radians(p2[1])
    
    # calcul des différences
    d_lat = lat2 - lat1
    d_lon = lon2 - lon1 
    
    Q = sqrt((sin(d_lat/2.))**2 + cos(lat1)*cos(lat2)*(sin(d_lon/2.))**2)
    dist = 2*6371*asin(Q)
    
    return dist
    
    
```

## Distribution géographique des arbres

### Ajout de colonnes Longitude et Latitude

```python
# display(dfGre['GeoJSON'].loc[24420])

LonLat = LONLATfromJSONlist(dfGre['GeoJSON'].tolist())
# ou : df = pd.read_csv('../data/arbres.csv', sep=";", low_memory=False)
# eval(df.loc[2, "GeoJSON"])['coordinates']
dfGre['Latitude'] = LonLat['LAT']
dfGre['Longitude'] = LonLat['LON']

display(dfGre[['Latitude','Longitude']].head())

geometry = geopandas.points_from_xy(dfGre['Longitude'],dfGre['Latitude'])
geodf = geopandas.GeoDataFrame(dfGre[['Latitude','Longitude']], geometry=geometry)
geodf.head()
print(len(geodf))

# Only keep 1000 arbres
geodf.sort_values(by=['Latitude','Longitude'])
print('Latitude between :',geodf['Latitude'].min(),geodf['Latitude'].max())
print('Longitude between :',geodf['Longitude'].min(),geodf['Longitude'].max())

mask = (geodf['Latitude'] > 45.17) & (geodf['Latitude'] < 45.18) & (geodf['Longitude'] > 5.74) & (geodf['Longitude'] < 5.8)
geodfPlot = geodf[mask].reset_index(drop=True)

display(geodfPlot.head())
print(len(geodfPlot))
```

### Création de Folium map

```python
map = folium.Map(location=[45.176, 5.745], tiles="OpenStreetMap", zoom_start=16) # /!\ ordre : latitude,longitude

geodfLi = [[point.xy[1][0], point.xy[0][0]] for point in geodfPlot['geometry']]
i=0
for coordinates in geodfLi:
    map.add_child(
        folium.Marker(
        location=coordinates, icon=folium.Icon(color='green')
        )
    )
map
```

### Matrice de distance entre 2 arbres

```python
arbre1 = (geodfPlot.loc[0,'geometry'].xy[1][0], geodfPlot.loc[0,'geometry'].xy[0][0])
arbre2 = (geodfPlot.loc[1000,'geometry'].xy[1][0], geodfPlot.loc[1000,'geometry'].xy[0][0])
print("La distance entre les 2 arbres est :", GRC(arbre1,arbre2).km, "km")
print("Avec la formule de Haversine, la distance est :", haversine(arbre1,arbre2), "km")

n = len(geodfPlot)
distMat = np.zeros(shape=(n,n))

for i, rowi in geodfPlot.iterrows():
    for j, rowj in geodfPlot.iterrows():
        if(j>i):
            arbre_i = (rowi['Latitude'],rowi['Longitude'])
            arbre_j = (rowj['Latitude'],rowj['Longitude'])
    
            distMat[i,j] = haversine(arbre_i, arbre_j)
        

print(distMat)
```

#### 10 arbres les plus proches 

```python
adressCoord = (45.176, 5.745)
distArbre = {}
for i, rowi in geodfPlot.iterrows():
    arbre_i = (rowi['Latitude'],rowi['Longitude'])
    distArbre[i] = haversine(arbre_i, adressCoord)
distArbre=dict(sorted(distArbre.items(), key=lambda x: x[1]))
indexNearest = [key for key in list(distArbre)[:10]]
nearestPlot = geodfPlot.loc[indexNearest]
display(nearestPlot)
```

#### Distance moyenne entre les arbres

```python
mask = np.full(shape=(n,n), fill_value=False)
for i in range(n):
    for j in range(n):
        mask[i,j] = (i<j)
print(mask)
print("La distance moyenne entre les arbres est de :", np.mean(distMat, where=mask)*1000, "m")
```

#### Arbre le plus entouré et le plus solitaire

```python
met = np.zeros(n)
for i in range(n):
    for j in range(n):
        if(i<j):
            met[i] += distMat[i,j]
        elif(i>j):
            met[i] += distMat[j,i]
    met[i] = sqrt(met[i])
```
